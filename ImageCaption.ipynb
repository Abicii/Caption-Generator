```python
import os 
from PIL import Image
from transformers import pipeline
```

```python
generator = pipeline(model="Salesforce/blip-image-captioning-base")
```

```python
img_path = os.path.join(os.getcwd(), 'Image3.png')
raw_image = Image.open(img_path).convert('RGB')
```

```python
output = generator(raw_image)
text_model_input = output[0]['generated_text']
print(text_model_input)
```

```python
def caption_generator(input):
  
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt= "write 5 instagram captions for an image described as" + "\n" + input,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )

    # Get the generated caption from the response
    captions = response["choices"][0]["text"]
    out = captions.split("\n")
    for caption in out:
      print(caption)
        
    
```

```python
caption_generator(text_model_input)
```
